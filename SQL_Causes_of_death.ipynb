```python
import pandas as pd

from config import username, password

from sqlalchemy import create_engine
pd.options.mode.chained_assignment = None
engine=create_engine(f"postgresql://{username}:{password}@localhost:5432/CausesofDeathProject")
connection=engine.connect()
```

```python
#read fianl leading causes of death csv
obesity = pd.read_csv('clean_datasets/clean_obesity.csv')
obesity
```

```python
#read fianl alcohol csv
alcohol = pd.read_csv('clean_datasets/clean_alcoholdata.csv')
alcohol
```

```python
#read fianl tobacco csv
tobacco = pd.read_csv('clean_datasets/clean_tobacco.csv')
tobacco
```

```python
#read fianl income csv
income = pd.read_csv('clean_datasets/final_cleaned_income_us.csv')
income
```

```python
deaths = pd.read_csv('clean_datasets/cleaned_leading_causes.csv')
deaths
```

```python
frames = [obesity, alcohol, tobacco, income, deaths]
merged_results = pd.concat(frames, axis=1, sort=False)
merged_results.to_csv(r'clean_datasets/merged_results.csv', index = False)
merged_results
```

```python
analysis=merged_results.drop(axis=0,columns=["alcohol_consumption"])
analysis.to_csv('Output/Analysis.csv')
analysis['per capita income'] = analysis['per capita income'].astype(int)
```

```python
#Load Dataframe into databases
analysis.to_sql(name='merged_results', con=engine, if_exists='replace', index=False)
```
