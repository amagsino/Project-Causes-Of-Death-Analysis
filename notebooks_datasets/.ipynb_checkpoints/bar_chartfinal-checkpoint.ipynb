```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objs as go
import mpl_toolkits.axisartist as AA
from mpl_toolkits.axes_grid1 import host_subplot

import seaborn as sns
```

```python
NJ_analysis_df = pd.read_csv('../clean_datasets/NJ_analysis_df.csv')
NJ_analysis_df
```

```python
state= NJ_analysis_df['state']
state

year= NJ_analysis_df['year']
year

per_capita_income= NJ_analysis_df['per capita income']
per_capita_income

tobacco_percentage= NJ_analysis_df['tobacco_percentage']
tobacco_percentage

obesity_percentage= NJ_analysis_df['obesity_percentage']
obesity_percentage
                                
```

```python
host = host_subplot(111, axes_class=AA.Axes)
plt.subplots_adjust(right=0.75)

par3 = host.twinx()
par2 = host.twinx()


offset = 60
new_fixed_axis = par2.get_grid_helper().new_fixed_axis
par2.axis["right"] = new_fixed_axis(loc="right",
                                    axes=par2,
                                    offset=(offset, 0))

par3.axis["right"].toggle(all=True)

host.set_xlim(2011, 2016)
host.set_ylim(40000, 65000)

host.set_xlabel("Year")
host.set_title("NJ Causes of Death")
host.set_ylabel("Per Capita Income")
par3.set_ylabel("Tobacco Use")
par2.set_ylabel("Obesity")

p1, = host.plot([2011,2012,2013,2014,2015,2016],[53696,55583,56045,58655,61110, 62701], label='Per Capita Income')
p2, = par3.plot([2011,2012,2013,2014,2015,2016],[18.8,16.8,22.2, 26.2, 15.1,19.1], label='Tobacco Use')
p3, = par2.plot([2011,2012,2013,2014,2015,2016],[23.81, 24.74, 26.39, 27.07, 25.69, 27.49], label='Obesity')

par3.set_ylim(0, 40)
par2.set_ylim(0, 40)

host.legend()

host.axis["left"].label.set_color(p1.get_color())
par3.axis["right"].label.set_color(p2.get_color())
par2.axis["right"].label.set_color(p3.get_color())

plt.draw()
plt.show()
plt.savefig('../bar_chart.png')
```
