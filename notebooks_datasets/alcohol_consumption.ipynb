```python
#Import depencies 
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#Read in CSV 
alcohol=('../original_datasets/alcohol_consumption.csv')

alcohol_consumption=pd.read_csv(alcohol)
alcohol_consumption
```

```python
alcohol_consumption.info()
```

```python
alcohol_consumption['year']= alcohol_consumption['year'].map("{:.0f}".format)

alcohol_consumption
```

```python
drop_column=alcohol_consumption.drop('all_beverages', axis=1)
drop_column.head()
```

```python
drop_column2=drop_column.drop('beer', axis=1)
drop_column2
```

```python
drop_column3=drop_column2.drop('wine', axis=1)
drop_column3
```

```python
drop_column4=drop_column3.drop('spirits', axis=1)
drop_column4
```

```python
drop_column5=drop_column4.drop('state', axis=1)
drop_column5
```

```python
drop_column6=drop_column5.drop('year', axis=1)
drop_column6
```

```python
#2011-2016
clean_alcoholdata=drop_column6
clean_alcoholdata
```

```python
clean_alcoholdata.reset_index(drop = True, inplace = True)
clean_alcoholdata.to_csv(r'../clean_datasets/clean_alcoholdata.csv', index = False)
```
