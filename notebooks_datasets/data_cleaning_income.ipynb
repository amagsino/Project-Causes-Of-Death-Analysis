```python
#Import depencies 
import pandas as pd
import numpy as np
```

```python
#Read in CSV 
income=('../original_datasets/per_cap_income_us.csv')

per_cap_income=pd.read_csv(income, header=4)
per_cap_income.head()
```

```python
#Drop NaN values 
per_cap_df_dropped=per_cap_income.apply(pd.to_numeric, errors='coerce')
per_cap_df_dropped=per_cap_income.dropna()
per_cap_df_dropped
```

```python
#Drop Row/Col
drop_col=per_cap_df_dropped.drop('GeoFips', axis=1)
drop_col.head()
```

```python
drop_col['Year']=drop_col['Year'].values.astype(np.int64)
drop_col
```

```python
drop_col.info()
```

```python
#Rename column
drop1_df=drop_col.drop('GeoName', axis=1)
drop1_df
```

```python
drop2_df=drop1_df.drop('Year', axis=1)
drop2_df
```

```python
col_rename=drop2_df.rename(columns={'Per Capita Income':'per capita income'})
col_rename['per capita income'] = col_rename['per capita income'].astype(int)
col_rename
```

```python
#Remove asterik
remove_ast=col_rename.replace('\*','',regex=True).astype(str)
remove_ast.reset_index(drop = True, inplace = True)
remove_ast
```

```python
remove_ast.to_csv(r'../clean_datasets/final_cleaned_income_us.csv', index = False)
```
