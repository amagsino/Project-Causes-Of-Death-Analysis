```python
#Import depencies 
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#Read in CSV 
cause_death=('../original_datasets/leading_causes_death_us.csv')

leading_causes=pd.read_csv(cause_death)
leading_causes
```

```python
df=pd.DataFrame(leading_causes)
df['Deaths'] = df['Deaths'].astype(int)
df.dtypes
```

```python
drop_col1=df.drop('113 Cause Name', axis=1)
drop_col1
```

```python
drop_col2=drop_col1.drop('Year', axis=1)
drop_col2
```

```python
drop_col3=drop_col2.drop('State', axis=1)
drop_col3
```

```python
drop_col4=drop_col3.drop('Age-adjusted Death Rate', axis=1)
drop_col4
```

```python
drop_col5=drop_col4.drop('Cause Name', axis=1)
drop_col5
```

```python
drop_col5.to_csv(r'../clean_datasets/cleaned_leading_causes.csv', index = False)
```
