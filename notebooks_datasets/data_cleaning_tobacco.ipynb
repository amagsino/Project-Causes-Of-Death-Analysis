```python
tobacco_use = pd.read_csv('../original_datasets/Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv')
tobacco_use.head()
```

```python
#iterate through the columns
for col in tobacco_use:
    print(col)
```

```python
tobacco_use_df = tobacco_use[["LocationDesc", "YEAR", "LocationAbbr", "TopicDesc", "MeasureDesc", "Data_Value", "Gender", "Race", "Age", "GeoLocation"]]
tobacco_use_df.head()
```

```python
tobacco_use_df.columns = ["state", "year", "abbr", "description", "measure", "value", "gender", "race", "age", "geolocation"]
tobacco_use_df.head()
```

```python
tobacco_use_df = tobacco_use_df.dropna()
tobacco_use_df.head()
```

```python
clean_tobacco_usedf = tobacco_use_df[(tobacco_use_df.description == 'Cigarette Use (Adults)') & (tobacco_use_df.measure == 'Current Smoking')]
clean_tobacco_usedf.head()
```

```python
clean_tobacco = clean_tobacco_usedf[(clean_tobacco_usedf.gender == 'Overall') & (clean_tobacco_usedf.race == 'All Races') & (clean_tobacco_usedf.age == 'All Ages')]
clean_tobacco.head()
```

```python
clean_tobacco_df = clean_tobacco[["state", "year", "value"]]
clean_tobacco_df.head()
```

```python
clean_tobacco_df.columns = ["state", "year", "tobacco_percentage"]
```

```python
clean_tobacco_df=clean_tobacco_df.drop('state', axis=1)
clean_tobacco_df=clean_tobacco_df.drop('year', axis=1)
clean_tobacco_df
```

```python
clean_tobacco_df.reset_index(drop = True, inplace = True)
clean_tobacco_df.to_csv(r'../clean_datasets/clean_tobacco.csv', index = False)
```
